```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
import pickle
```

```python
monet_image_file_names = os.listdir('monet_jpg')
monet_images = np.array([ img.imread('monet_jpg/' + file_name) / 255.0 for file_name in monet_image_file_names ], dtype=np.float32)
with open("monet_image_file_names.pickle", "wb") as f:
    pickle.dump(monet_image_file_names, f)
with open("monet_images.pickle", "wb") as f:
    pickle.dump(monet_images, f, protocol=4)
```

```python
plt.imshow(monet_images[0])
```

```python
photo_image_file_names = os.listdir('photo_jpg')
photo_images = np.array([ img.imread('photo_jpg/' + file_name) / 255.0 for file_name in photo_image_file_names ], dtype=np.float32)
with open("photo_image_file_names.pickle", "wb") as f:
    pickle.dump(photo_image_file_names, f)
with open("photo_images.pickle", "wb") as f:
    pickle.dump(photo_images, f, protocol=4)
```
